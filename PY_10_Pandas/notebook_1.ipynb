Напишите функцию get_experience(arg), аргументом которой является строка столбца с опытом работы. Функция должна возвращать опыт работы в месяцах. Не забудьте привести результат к целому числу.

Примечание. Обратите внимание, что опыт работы может выражаться только в годах или только в месяцах. Учтите это при построении своей функции.

```python
import pandas as pd
hh = pd.Series(data = ['опыт работы 12 лет 3 месяца', 'опыт работы 3 года 11 месяцев', 'опыт работы 3 года', 'опыт работы 3 месяца'])
display(hh)

def get_experience(arg): 
    splited_arg = arg.split(' ')
    digits_only = []
    for x in splited_arg:
        if x.isdigit():
            digits_only.append(int(x))
    if len(digits_only) == 2:
        months_count = digits_only[0]*12 + digits_only[1]
    elif splited_arg[-1] in ['лет', 'года']:
        months_count = digits_only[0]*12
    elif splited_arg[-1] in ['месяцев', 'месяца']:
        months_count = digits_only[0]
    return months_count

hh.apply(get_experience)

```
